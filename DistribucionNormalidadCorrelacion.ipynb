```python
import numpy as np
import pandas as pd

# Supongamos que los datos ya están cargados en la variable 'datos'
data = pd.read_csv('/content/insurance.csv')
datos = data['charges'].values
n = len(datos)  # Tamaño de la muestra
rango_datos = np.max(datos) - np.min(datos)

# Regla de Sturges
k_sturges = int(np.ceil(np.log2(n) + 1))

# Regla de la Raíz Cuadrada
k_raiz_cuadrada = int(np.ceil(np.sqrt(n)))

# Regla de Scott
sigma = np.std(datos)
k_scott = int(np.ceil(rango_datos / (3.5 * sigma / (n**(1/3)))))

# Regla de Freedman-Diaconis
Q1 = np.percentile(datos, 25)
Q3 = np.percentile(datos, 75)
IQR = Q3 - Q1
k_freedman_diaconis = int(np.ceil(rango_datos / (2 * IQR / (n**(1/3)))))

# Mostrar el número de intervalos sugeridos por cada método
print(f"Número de intervalos según la regla de Sturges: {k_sturges}")
print(f"Número de intervalos según la regla de la Raíz Cuadrada: {k_raiz_cuadrada}")
print(f"Número de intervalos según la regla de Scott: {k_scott}")
print(f"Número de intervalos según la regla de Freedman-Diaconis: {k_freedman_diaconis}")
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats

# Leer el archivo CSV (ajusta la ruta si es necesario)
data = pd.read_csv('/content/insurance.csv')
var_dependiente = data['charges'].values
n = len(var_dependiente)
#Calculamos las estadisticas descriptivas
#Media
media = np.sum(var_dependiente/len(var_dependiente))

#Mediana
var_ordenada = np.sort(var_dependiente)
if n % 2 == 0:
    mediana = ((var_ordenada[(n//2)-1])+var_ordenada[(n//2)])/2
else:
    mediana = var_ordenada[n//2]

#Varianza
varianza = sum((x - media) ** 2 for x in var_dependiente)/n

#Desviacion estandar
desviacion = np.sqrt(varianza)

#Obtenemos el histograma

# Calculamos cuantos bins hacer
rango_datos = np.max(var_dependiente) - np.min(var_dependiente)
Q1 = np.percentile(var_dependiente, 25)
Q3 = np.percentile(var_dependiente, 75)
IQR = Q3 - Q1
bins = int(np.ceil(rango_datos / (2 * IQR / (n**(1/3)))))

#Definimos los limites
intervalos = np.linspace(np.min(var_dependiente),np.max(var_dependiente),bins+1)

#Obtenemos las frecuencias
frecuencias = []
for i in range(len(intervalos) - 1 ):
  #Contamos cuantos hay dentro de cada intervalo
  frecuencia = np.sum((datos >= intervalos[i]) & (datos < intervalos[i+1]))
  frecuencias.append(frecuencia)
#Para el ultimo contamos los que si son exactos al maximo
frecuencias[-1] += np.sum(datos == np.max(var_dependiente))

for i in range(len(frecuencias)):
    print(f'Intervalo {intervalos[i]:.2f} - {intervalos[i+1]:.2f}: {frecuencias[i]}')

#Graficamos
plt.bar(intervalos[:-1],frecuencias, width=np.diff(intervalos),align='edge',edgecolor = 'black')
# Etiquetas, título
plt.xlabel('Intervalos')
plt.ylabel('Frecuencia')
plt.title('Histograma')

# Mostrar el gráfico
plt.show()

# Crear un gráfico Q-Q para verificar la normalidad
plt.figure(figsize=(8, 6))
stats.probplot(data['charges'], dist="norm", plot=plt)
plt.title('Gráfico Q-Q para Verificar Normalidad de Charges', fontsize=15)
plt.show()


#Los descriptivos
print(data['charges'].describe())
print(f"Min: {np.min(var_dependiente)}")
print(f"Q1: {Q1}")
print(f"Mediana: {mediana}")
print(f"Q3: {Q3}")
print(f"Max: {np.max(var_dependiente)}")
print(f"Media: {media}")
print(f"Varianza: {varianza}")
print(f"Desviación estándar: {desviacion}")
```
